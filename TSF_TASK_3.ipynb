##### Data Visualization

```python
# visualizing the data as a whole using the pair plot
sns.pairplot(df)
```

```python
#Analyzing relationships between variables using correlation matrix

df.corr()
plt.figure(figsize=(12,6))
sns.heatmap(df.corr(), annot = True, cmap=sns.diverging_palette(170, 10, as_cmap=True))
```

```python
plt.figure(figsize=(5,5))
plt.title('Ship Modes')
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
explode = (0.1, 0, 0.1, 0)
plt.pie(df['Ship Mode'].value_counts(), explode=explode, labels=df['Ship Mode'].value_counts().index, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.title('Sub-Category')
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(df['Category'].value_counts(), colors=colors, labels=df['Category'].value_counts().index,autopct='%1.1f%%', startangle=90 )
centre_circle = plt.Circle((0,0),  0.70, fc = 'white')
fig= plt.gcf()
fig.gca().add_artist(centre_circle)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (15,8))
sns.countplot(x = df['State'])
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize = (6,6))
textprops = 
plt.title('Sales vs Profit in Different Regions')
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(df['Region'].value_counts(),colors=colors, labels=df['Region'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

```python
# computing top states in terms of sales from first 10 observations
top_states_s = df.groupby("State").Sales.sum().nlargest(n=10)

# computing top states in terms of profit from first 10 observations
top_states_p = df.groupby("State").Profit.sum().nlargest(n = 10)

plt.style.use('seaborn')
top_states_s.plot(kind = 'bar',figsize = (10,5),fontsize = 14, color = '#ff9999')
top_states_p.plot(kind = 'bar',figsize = (10,5),fontsize = 14, color = '#ffcc99')
plt.xlabel('States',fontsize = 15)
plt.ylabel('Total sales',fontsize = 15)
plt.title("Top 10 states Sales vs Profit",fontsize = 15)
plt.show()
```

##### Checking the interdependency of Sales,Profits and Discounts

```python
plt.style.use('seaborn')
df.plot(kind = "scatter",figsize = (10,5), x = "Sales", y= "Profit", c = "Discount", s = 20,fontsize = 16, colormap = "viridis")
plt.ylabel('Total Profits',fontsize = 16)
plt.title("Interdependency of Sales,Profits and Discounts",fontsize = 16)

plt.show()
```
